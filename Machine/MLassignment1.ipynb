```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('C:\\Users\\msi\\Desktop\\ML\\cars.csv')
df.isnull().any()
missingValues=['horsepower', 'origin']

#find the number of missing value in each one
for x in missingValues:
    counter=df[x].isnull().sum()
    print(f"number of missing values in {x}")
    print(counter)

# fill the missing value in origin using mode
fillOrigin='origin'
modeValue=df[fillOrigin].mode().iloc[0]
print('Mode Value',modeValue)
df[fillOrigin].fillna(modeValue,inplace=True)

#to check if there are outliers 
upperLimit= df['horsepower'].mean() + 3*df['horsepower'].std()
lowerLimit= df['horsepower'].mean() - 3*df['horsepower'].std()
print('upper limit', upperLimit)
print('lower limit', lowerLimit)
df.loc[(df['horsepower'] > upperLimit) | (df['horsepower']< lowerLimit)]

#fill the missing values using median
medianValue=df['horsepower'].median()
print('Median Value ',medianValue)
df['horsepower'].fillna(medianValue,inplace=True)

x=df['mpg'].values
y=df['horsepower'].values
X = np.column_stack((np.ones_like(x), x))#x0=1
coefficients = np.linalg.inv(X.T @ X) @ X.T @ y
beta_0, beta_1 = coefficients
plt.scatter(x ,y,label='Scatter Plot')
plt.plot(x, beta_0 + beta_1 * x, color='green', label='Linear Regression Line')
plt.title('Linear Regression for mpg from horsepower')
plt.xlabel('horsepower')
plt.ylabel('mpg')
plt.legend()
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('C:\\Users\\msi\\Desktop\\ML\\cars.csv')
df.isnull().any()
missingValues=['horsepower', 'origin']

#find the number of missing value in each one
for x in missingValues:
    counter=df[x].isnull().sum()
    print(f"number of missing values in {x}")
    print(counter)

# fill the missing value in origin using mode
fillOrigin='origin'
modeValue=df[fillOrigin].mode().iloc[0]
print('Mode Value',modeValue)
df[fillOrigin].fillna(modeValue,inplace=True)

#to chrck if there are outliers 
upperLimit= df['horsepower'].mean() + 3*df['horsepower'].std()
lowerLimit= df['horsepower'].mean() - 3*df['horsepower'].std()
print('upper limit', upperLimit)
print('lower limit', lowerLimit)
df.loc[(df['horsepower'] > upperLimit) | (df['horsepower']< lowerLimit)]

#fill the missing values using median
medianValue=df['horsepower'].median()
print('Median Value ',medianValue)
df['horsepower'].fillna(medianValue,inplace=True)

x=df['mpg'].values
y=df['horsepower'].values
X = np.column_stack((np.ones_like(x), x, x**2))
coefficients = np.linalg.inv(X.T @ X) @ X.T @ y
w0, w1, w2 = coefficients
# Create a scatter plot
plt.scatter(x, y, label='Scatter Plot')
x_range = np.linspace(min(x), max(x), 100)
y_quad = w0 + w1 * x_range + w2 * (x_range**2)
plt.plot(x_range, y_quad, color='green', label='Quadratic Regression')
plt.xlabel('horsepower')
plt.ylabel('mpg')
plt.title('Quadratic Regression for mpg and horsepower')
plt.legend()
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('C:\\Users\\msi\\Desktop\\ML\\cars.csv')
df.isnull().any()
missingValues=['horsepower', 'origin']

#find the number of missing value in each one
for x in missingValues:
    counter=df[x].isnull().sum()

# fill the missing value in origin using mode
fillOrigin='origin'
modeValue=df[fillOrigin].mode().iloc[0]
df[fillOrigin].fillna(modeValue,inplace=True)

#to chrck if there are outliers 
upperLimit= df['horsepower'].mean() + 3*df['horsepower'].std()
lowerLimit= df['horsepower'].mean() - 3*df['horsepower'].std()
df.loc[(df['horsepower'] > upperLimit) | (df['horsepower']< lowerLimit)]

#fill the missing values using median
medianValue=df['horsepower'].median()
df['horsepower'].fillna(medianValue,inplace=True)

x=df['mpg'].values
y=df['horsepower'].values
learning_rate = 0.001 
th0 = 0.0
th1 = 0.0
epochs = 199999
numOfData = len(x)
for epoch in range(epochs):

    h = th0 + th1 * x
    gradient_0 = (1/numOfData) * np.sum(h - y)
    gradient_1 = (1/numOfData) * np.sum((h - y) * x)
    th0 -= learning_rate * gradient_0
    th1 -= learning_rate * gradient_1

plt.scatter(x, y, label='Scatter Plot')
plt.plot(x, th0 + th1 * x, color='pink', label='Gradient Descent ')
plt.title('Gradient Descent for Predicting mpg from horsepower')
plt.xlabel('horsepower')
plt.ylabel('mpg')
print('theta0-> {:.2f}'.format(th0))
print('theta1-> {:.2f}'.format(th1))
plt.legend()
plt.show()
```
